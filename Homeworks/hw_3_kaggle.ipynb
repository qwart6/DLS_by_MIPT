```python
num_cols = [
    'ClientPeriod',
    'MonthlySpending',
    'TotalSpent'
]

cat_cols = [
    'Sex',
    'IsSeniorCitizen',
    'HasPartner',
    'HasChild',
    'HasPhoneService',
    'HasMultiplePhoneNumbers',
    'HasInternetService',
    'HasOnlineSecurityService',
    'HasOnlineBackup',
    'HasDeviceProtection',
    'HasTechSupportAccess',
    'HasOnlineTV',
    'HasMovieSubscription',
    'HasContractPhone',
    'IsBillingPaperless',
    'PaymentMethod'
]

feature_cols = num_cols + cat_cols
target_col = 'Churn'
```

#    Totalspent - object,     

```python
for col in num_cols:
    data_train[col] = pd.to_numeric(data_train[col], errors='coerce')
    data_test[col]  = pd.to_numeric(data_test[col], errors='coerce')
```

```python
nun_info = pd.DataFrame(
    data = {
        'type': ['numeric' if col in num_cols else 'categorical' for col in feature_cols],
        'Train' : data_train.drop(columns='Churn')[feature_cols].isna().sum(),
        'Test' : data_test[feature_cols].isna().sum()
})

nun_info = nun_info.sort_values('type')
nun_info['type'] = nun_info['type'].mask(nun_info['type'].duplicated(), '  ')
nun_info
```
